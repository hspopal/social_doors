```python
import os

from nilearn.maskers import NiftiMasker
from nltools.data import Brain_Data
from nilearn.plotting import plot_glass_brain, plot_stat_map

from nilearn.image import math_img, smooth_img
```

```python
bids_dir = '/data/projects/social_doors/'
os.chdir(bids_dir)

data_dir = os.path.join(bids_dir, 'derivatives', 'rois')
```

```python
quarmley_VS_L = Brain_Data(data_dir+'/VS_L.nii', 
                  mask=bids_dir+"derivatives/fmriprep/sub-010/func/sub-010_task-mdoors_run-1_space-MNI152NLin2009cAsym_desc-brain_mask.nii.gz")
```

```python
quarmley_VS_L = quarmley_VS_L.threshold(upper=0.1, binarize=True)

quarmley_VS_L.to_nifti().to_filename(data_dir+'/VS_L.nii')

plot_glass_brain(quarmley_VS_L.to_nifti())
```

```python
# Import reward neurosynth map

reward_map = Brain_Data(data_dir+'/reward_association-test_z_FDR_0.01.nii.gz',
                       mask=bids_dir+"derivatives/fmriprep/sub-010/func/sub-010_task-mdoors_run-1_space-MNI152NLin2009cAsym_desc-brain_mask.nii.gz")

# Smooth
reward_map = reward_map.smooth(4)

# Threshold
reward_map = reward_map.threshold(upper=5)

plot_glass_brain(reward_map.to_nifti())
```

```python
# Binarize
reward_map_bin = reward_map.threshold(upper=5, binarize=True)

reward_rois = reward_map_bin.regions(min_region_size=200)

for n in range(len(reward_rois)):
    plot_stat_map(reward_rois[n].to_nifti(), 
                 cut_coords=range(-65,66,10), display_mode='x',
                 cmap='Reds')
```

```python
reward_roi_names = ['VS_l','vmPFC_b','VS_r']

reward_rois_dict = {}

for n in range(len(reward_rois)):
    reward_rois_dict[reward_roi_names[n]] = reward_rois[n]
    reward_rois_dict[reward_roi_names[n]].to_nifti().to_filename(data_dir+'/roi_bin_reward_'+reward_roi_names[n]+'.nii.gz')
    
```

# Reinforcement Learning ROIs

```python
rlearn_map = Brain_Data(data_dir+'/reinforcement-learning_association-test_z_FDR_0.01.nii.gz',
                       mask=bids_dir+"derivatives/fmriprep/sub-010/func/sub-010_task-mdoors_run-1_space-MNI152NLin2009cAsym_desc-brain_mask.nii.gz")

# Smooth
rlearn_map = rlearn_map.smooth(4)

# Threshold
rlearn_map = rlearn_map.threshold(upper=2)

plot_glass_brain(rlearn_map.to_nifti())
```

```python
# Binarize
rlearn_map_bin = rlearn_map.threshold(upper=2, binarize=True)

rlearn_rois = rlearn_map_bin.regions(min_region_size=50)

for n in range(len(rlearn_rois)):
    plot_stat_map(rlearn_rois[n].to_nifti(), 
                 cut_coords=range(-65,66,5), display_mode='x',
                 cmap='Reds')
```

```python
rlearn_roi_names = ['VS_l','vmPFC_r','LIX_r','VS_r','C1_r']

rlearn_rois_dict = {}

for n in range(len(rlearn_rois)):
    rlearn_rois_dict[rlearn_roi_names[n]] = rlearn_rois[n]
    rlearn_rois_dict[rlearn_roi_names[n]].to_nifti().to_filename(data_dir+'/roi_bin_reinforcement-learning_'+rlearn_roi_names[n]+'.nii.gz')
    
```

# Social ROIs

```python
social_map = Brain_Data(data_dir+'/social_association-test_z_FDR_0.01.nii.gz',
                       mask=bids_dir+"derivatives/fmriprep/sub-010/func/sub-010_task-mdoors_run-1_space-MNI152NLin2009cAsym_desc-brain_mask.nii.gz")

# Smooth
social_map = social_map.smooth(4)

# Threshold
social_map = social_map.threshold(upper=2)

plot_glass_brain(social_map.to_nifti())
```

```python
# Binarize
social_map_bin = social_map.threshold(upper=2, binarize=True)

social_rois = social_map_bin.regions(min_region_size=200)

for n in range(len(social_rois)):
    plot_stat_map(social_rois[n].to_nifti(), 
                 cut_coords=range(-65,66,10), display_mode='x',
                 cmap='Reds')
```

```python
social_roi_names = ['TPJ_l', 'ATL_l', 'PCC_b', 'dmPFC_b', 'AMG_r', 'C1_r',
                    'TPJ_r', 'ATL_r']

social_rois_dict = {}

for n in range(len(social_rois)):
    social_rois_dict[social_roi_names[n]] = social_rois[n]
    social_rois_dict[social_roi_names[n]].to_nifti().to_filename(data_dir+'/roi_bin_social_'+social_roi_names[n]+'.nii.gz')
    
```

```python
social_rois_dict['TPJ_b'] = social_rois_dict['TPJ_l'] + social_rois_dict['TPJ_r']
social_rois_dict['ATL_b'] = social_rois_dict['ATL_l'] + social_rois_dict['ATL_r']

social_rois_dict['TPJ_b'].to_nifti().to_filename(data_dir+'/roi_bin_social_TPJ_b.nii.gz')
social_rois_dict['ATL_b'].to_nifti().to_filename(data_dir+'/roi_bin_social_ATL_b.nii.gz')
```

```python
from nilearn import datasets
from nilearn import surface


fsaverage = datasets.fetch_surf_fsaverage()

texture = surface.vol_to_surf(social_map.to_nifti(), fsaverage.pial_left)

from nilearn import plotting

fig = plotting.plot_surf_stat_map(
    fsaverage.infl_left, texture, hemi='left',
    title='Surface left hemisphere', colorbar=True,
    threshold=1., bg_map=fsaverage.sulc_left
)
fig.show()
```

```python
plotting.plot_img_on_surf(social_map.to_nifti(),
                          views=['lateral', 'medial'],
                          hemispheres=['left', 'right'],
                          inflate=True,
                          threshold=1,
                         surf_mesh='fsaverage')
plotting.show()
```

```python
from nilearn import image
import SUITPy as suit
import SUITPy.flatmap as flatmap

c1_social_mask = 1 * (image.get_data(social_rois_dict['C1_r'].to_nifti()) > 0.01)
c1_rlearn_mask = 2 * image.get_data(rlearn_rois_dict['C1_r'].to_nifti())
LIX_rlearn_mask = 2 * image.get_data(rlearn_rois_dict['LIX_r'].to_nifti())

combined_mask = image.new_img_like(social_rois_dict['C1_r'].to_nifti(),
                                  c1_social_mask + c1_rlearn_mask + LIX_rlearn_mask)


cb_rois = flatmap.vol_to_surf(combined_mask)

flatmap.plot(data=cb_rois, cmap='tab10', threshold=[0,3], render='matplotlib',
    new_figure=True)
```
